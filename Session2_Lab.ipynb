```python
# 1.Import libraries and load the CSV

import pandas as pd
import matplotlib.pyplot as plt

# (Optional) Ensure plots appear inline if you’re using a Jupyter notebook
# %matplotlib inline

# 1.1 Read the CSV into a DataFrame
df = pd.read_csv("cs_Dataset.csv")

# 1.2 Inspect the first few rows
print(df.head())
```

**Explanation**

pd.read_csv("your_dataset.csv") loads the CSV into a DataFrame called df.

df.head() shows the first 5 rows, so you can verify column names and a few sample values

```python
# 2. Inspect DataFrame structure
# 2.1 See column names, non-null counts and dtypes
df.info()

# 2.2 Summary statistics for all numerical columns
print(df.describe())

# 2.3 Count unique values in categorical columns (if any)
for col in df.select_dtypes(exclude="number").columns:
    print(f"\nColumn '{col}' value counts:")
    print(df[col].value_counts())
```

**Explanation**


* df.info() tells you how many non-null entries each column has, plus each column’s data type.
* df.describe() prints count, mean, std, min, max and quartiles for every numerical column.
* If you have columns that are strings (e.g. "protocol_type", "flag", "label"), using value_counts() on them helps you see how many times each category appears.








```python
# 3. Visualise key numerical features
# Assuming your dataset has numerical columns such as duration, src_bytes and dst_bytes, you can plot histograms to see their distributions:

# 3.1 List the numerical columns you want to plot
numerical_cols = ["duration", "src_bytes", "dst_bytes"]  # adjust to your actual column names

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    df[col].hist(bins=10)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

**Explanation**

* This creates one histogram per numerical feature.

* Adjust bins=10 (or any integer) to change granularity

```python
# 4. Visualise categorical‐feature distributions

# If you have categorical columns like "label" (e.g. “normal” vs. “intrusion”) or "protocol_type", "flag", you can use bar charts:


# 4.1 Plot the 'label' distribution
plt.figure(figsize=(5, 3))
label_counts = df["label"].value_counts()
label_counts.plot(kind="bar", color=["green", "red"])
plt.title("Label Distribution")
plt.xlabel("Label")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 4.2 Plot 'protocol_type' distribution (if present)
if "protocol_type" in df.columns:
    plt.figure(figsize=(5, 3))
    protocol_counts = df["protocol_type"].value_counts()
    protocol_counts.plot(kind="bar", color="skyblue")
    plt.title("Protocol Type Distribution")
    plt.xlabel("Protocol Type")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# 4.3 Plot 'flag' distribution (if present)
if "flag" in df.columns:
    plt.figure(figsize=(5, 3))
    flag_counts = df["flag"].value_counts()
    flag_counts.plot(kind="bar", color="orange")
    plt.title("Flag Distribution")
    plt.xlabel("Flag")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

**Explanation**

* Replace column names if yours differ.

* *These* bar plots let students quickly see how many records fall into each category.

```python
# 5. Scatter-plot relationships between two numeric features

# A scatter plot is helpful to see if there’s any pattern between, say, src_bytes and dst_bytes. We can also colour‐code points by the label column:

plt.figure(figsize=(6, 4))

# Map labels to colours
colour_map = {"normal": "blue", "intrusion": "red"}  # adjust values to match your dataset
point_colours = df["label"].map(colour_map)

plt.scatter(df["src_bytes"], df["dst_bytes"], c=point_colours, alpha=0.7)
plt.title("src_bytes vs. dst_bytes (coloured by label)")
plt.xlabel("src_bytes")
plt.ylabel("dst_bytes")
plt.tight_layout()
plt.show()
```

**Explanation**
* .map(colour_map) produces a colour for each row based on its label.

* **Blue** points might represent “normal”, red for “intrusion” (or however your data is labelled).

```python
# 6. Putting it all together
"""
 Below is a complete code snippet you can run in one go (e.g. in a single Python script or Jupyter cell).
 Just be sure to change "your_dataset.csv" and column names as needed

"""

import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the CSV
df = pd.read_csv("cs_Dataset.csv")

# 2. Inspect structure
print("=== DataFrame Info ===")
df.info()
print("\n=== First 5 Rows ===")
print(df.head())

print("\n=== Summary Statistics (Numerical) ===")
print(df.describe())

# 3. Categorical value counts (if any)
for col in df.select_dtypes(exclude="number").columns:
    print(f"\nColumn '{col}' value counts:")
    print(df[col].value_counts())

# 4. Histograms of key numerical features
numerical_cols = ["duration", "src_bytes", "dst_bytes"]  # adjust as needed
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    df[col].hist(bins=10)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

# 5. Bar charts for categorical features
if "label" in df.columns:
    plt.figure(figsize=(5, 3))
    df["label"].value_counts().plot(kind="bar", color=["green", "red"])
    plt.title("Label Distribution")
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

if "protocol_type" in df.columns:
    plt.figure(figsize=(5, 3))
    df["protocol_type"].value_counts().plot(kind="bar", color="skyblue")
    plt.title("Protocol Type Distribution")
    plt.xlabel("Protocol Type")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

if "flag" in df.columns:
    plt.figure(figsize=(5, 3))
    df["flag"].value_counts().plot(kind="bar", color="orange")
    plt.title("Flag Distribution")
    plt.xlabel("Flag")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# 6. Scatter plot: src_bytes vs. dst_bytes coloured by label
if {"src_bytes", "dst_bytes", "label"}.issubset(df.columns):
    plt.figure(figsize=(6, 4))
    colour_map = {"normal": "blue", "intrusion": "red"}
    plt.scatter(df["src_bytes"], df["dst_bytes"],
                c=df["label"].map(colour_map), alpha=0.7)
    plt.title("src_bytes vs. dst_bytes (coloured by label)")
    plt.xlabel("src_bytes")
    plt.ylabel("dst_bytes")
    plt.tight_layout()
    plt.show()

```

# In you are interested to know about flag.

In many network‐intrusion datasets (e.g. KDD’99 or NSL-KDD), the flag attribute encodes the status of a TCP connection using a two-letter abbreviation. These flags reflect how the TCP handshake and teardown proceeded. Here are some of the most common ones you’ll encounter:

SF (“SYN-FIN”):

The connection was successfully established (via a SYN→SYN-ACK→ACK handshake) and then cleanly closed with a FIN.

In other words, a normal connection that opened and closed without errors.

S0 (“SYN-no reply”):

A SYN packet was sent to initiate a connection, but no reply (SYN-ACK or RST) was received.

This often indicates a port scan or a “half-open” attempt that never completed.

S1 (“SYN-ESTABLISHED-no FIN”):

The SYN→SYN-ACK→ACK handshake was completed, so the connection was established, but it did not terminate via FIN.

In practice, this might signal a connection that remains open or was abruptly dropped on one side.

REJ (“REJECTED”):

A SYN packet was sent and the target explicitly responded with a TCP RST (reset), rejecting the connection.

Typically indicates an attempt to connect to a closed port or a service that refused the session.

S2 / S3 (less common):

S2: SYN→SYN-ACK handshake succeeded, but the connection was reset by the originator (no FIN).

S3: SYN→SYN-ACK handshake succeeded, but the connection was reset by the responder (no FIN).

SH (“SYN-RST-FIN”):

A connection that was established normally (SYN handshake) but then both sides sent a reset (RST) or FIN in quick succession.

Usually indicates an abnormal or forced termination
